## Part 2: Data Cleaning and Preparation

In this section, I'll focus on preparing the datasets for analysis by cleaning and formatting them appropriately. This will involve removing unnecessary columns and addressing missing data. Additionally, for the school incidents dataset, I will add a new column to indicate whether there were any victims injured or killed, which will aid in further analysis of the severity of incident

### Data Cleaning Steps

1. **Drop Unnecessary Columns**: Remove columns related to suspects and operationsinceas they are not relevant ttheuspecific r analysis goa that I havels.
2. **Add Indicator Column**: For the school incidents dataset, add a new column 'Victim Impact' that indicates whether there were any victims injured or killed.
3. **Export Cleaned Data**: Save the cleaned datasets back to the project folder for further analysiRunningcuting the following code will perform these steps and provide a preview of the cleaned data.s.

```python
import pandas as pd

# Load the datasets
mass_shootings = pd.read_csv('Mass Shootings 2023.csv')
school_incidents = pd.read_csv('School Incidents 2023.csv')

# Dropping unnecessary columns related to suspects
columns_to_drop = ['Suspects Killed', 'Suspects Injured', 'Suspects Arrested', 'Operations']
mass_shootings.drop(columns=columns_to_drop, inplace=True, errors='ignore')
school_incidents.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# Add a new column 'Victim Impact' to indicate if there were any victims injured or killed
school_incidents['Victim Impact'] = school_incidents['Victims Killed'] + school_incidents['Victims Injured']
school_incidents['Victim Impact'] = school_incidents['Victim Impact'].apply(lambda x: 'Yes' if x > 0 else 'No')

# Display the cleaned datasets
print("Cleaned Mass Shootings Dataset Preview:")
print(mass_shootings.head())
print("\nCleaned School Incidents Dataset Preview:")
print(school_incidents.head())


# Export the cleaned datasets
# mass_shootings.to_csv('cleaned_mass_shootings_2023.csv', index=False)
# school_incidents.to_csv('cleaned_school_incidents_2023.csv', index=False)
```

### **Monthly Trends: Mass Shootings**

First, I’ll analyze how mass shootings are distributed across months. I’ll calculate the total number of incidents, fatalities, and injuries for each month and visualize the trends to identify when these events are most frequent and sever

In this code, I'm analyzing the mass shootings data by grouping incidents by their month. The goal is to see how these incidents are distributed throughout the year and identify any trends. First, I extract the month from the 'Incident Date' column, group the data by month, and calculate the total number of incidents, fatalities, and injuries for each month. I also ensure that the months are displayed in chronological order for better readability. Finally, I display the resulting table to get a clear view of the monthly distribution of mass shootings.e.
.


```python
# Extract the month from the 'Incident Date' column and group data by it
monthly_mass_shootings = (
    mass_shootings
    .assign(Month=mass_shootings['Incident Date'].str[:3])  # Get the first three letters of the month
    .groupby('Month')
    .agg(
        Total_Mass_Shootings=('Incident ID', 'count'),
        Total_Killed=('Victims Killed', 'sum'),
        Total_Injured=('Victims Injured', 'sum')
    )
    .reset_index()
)

# Sort the data by month in chronological order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_mass_shootings['Month'] = pd.Categorical(monthly_mass_shootings['Month'], categories=month_order, ordered=True)
monthly_mass_shootings = monthly_mass_shootings.sort_values('Month')

# Display the monthly mass shootings data
monthly_mass_shootings

```

Now, I’ll visualize the total number of mass shootings per month using a bar chart. This visualization will help highlight which months experienced the highest and lowest number of mass shootings, giving a clearer picture of any seasonal or monthly patterns.

This code creates a bar chart to visualize the total number of mass shootings for each month. The data is grouped by month, and the chart makes it easy to compare monthly trends at a glance.

```python
import plotly.express as px

# Create a bar chart for total mass shootings per month
fig = px.bar(
    monthly_mass_shootings,
    x='Month',
    y='Total_Mass_Shootings',
    labels={'Total_Mass_Shootings': 'Number of Mass Shootings', 'Month': 'Month'},
    title="Total Mass Shootings Per Month"
)
fig.update_layout(
    xaxis_title="Month",
    yaxis_title="Number of Mass Shootings",
    bargap=0.2
)
fig.show()
```

The bar chart above shows the monthly distribution of mass shootings for the year. From the data, it is clear that certain months had significantly higher numbers of incidents. For example, May and July stand out as the months with the highest number of mass shootings, with over 70 mass shootings each. On the other hand, February, March, November, and December had fewer mass shootings, all below 50.

These trends may suggest seasonal patterns or external factors influencing the frequency of mass shootings. This analysis provides a foundation for understanding when these incidents are most likely to occur, which could possibly be helpful for policy makers and others, in preventative strategies.

Now, I'll create a line chart to visualize the total number of fatalities and injuries caused by mass shootings for each month. This visualization helps pinpoint when the highest levels of human impact occur, making it easier to identify patterns or trends in severity over the year.

This code generates a line chart to display the total number of fatalities and injuries caused by mass shootings for each month. It adds separate lines for 'Total Killed' and 'Total Injured' to make it clear which months saw the highest impact.


```python
# Create a line chart for total killed and injured per month

import plotly.graph_objects as go

fig = go.Figure()

# Add line for total killed
fig.add_trace(go.Scatter(
    x=monthly_mass_shootings['Month'],
    y=monthly_mass_shootings['Total_Killed'],
    mode='lines+markers',
    name='Total Killed'
))

# Add line for total injured
fig.add_trace(go.Scatter(
    x=monthly_mass_shootings['Month'],
    y=monthly_mass_shootings['Total_Injured'],
    mode='lines+markers',
    name='Total Injured'
))

# Update layout
fig.update_layout(
    title='Total Killed and Injured Per Month',
    xaxis_title='Month',
    yaxis_title='Number of Victims',
    legend_title='Victim Types',
    hovermode='x'
)

fig.show()
```

**Patterns and Observations:**

- Injuries (red line) consistently outnumber fatalities (blue line) every month, as shown clearly in the graph.
- Both fatalities and injuries steadily increase from January (aside from dropping a bit in February), peaking during the summer months (May, June, and July), before decreasing sharply in the fall (with the exception of a rise in October).
- The lower amount of fatalities and injuries in February, March, November, and December, indicates a tapering off of mass shooting severity towards the beginning and end of the year.
- The summer months (May through July) represent the most severe period when it comes to injuries, with the highest injuries recorded.
- The sharp drop-off in injuries from July (400) to August (237) is a notable transition, suggesting a shift in incident patterns or severity.

```python
# Aggregate data by day of the week
mass_shootings['Day_of_Week'] = pd.to_datetime(mass_shootings['Incident Date']).dt.day_name()
mass_shootings_weekly = (
    mass_shootings
    .groupby('Day_of_Week')
    .agg(
        Total_Incidents=('Incident ID', 'count'),
        Total_Killed=('Victims Killed', 'sum'),
        Total_Injured=('Victims Injured', 'sum')
    )
    .sort_values(by='Total_Incidents', ascending=False)
    .reset_index()
)

# Get the maximum value for the y-axis range and calculate buffer
max_value = mass_shootings_weekly['Total_Incidents'].max()

# Bar chart for weekly trends in mass shootings (sorted with values displayed)
fig = px.bar(
    mass_shootings_weekly,
    x='Day_of_Week',
    y='Total_Incidents',
    title='Mass Shootings by Day of the Week (Most to Least)',
    labels={'Total_Incidents': 'Number of Mass Shootings', 'Day_of_Week': 'Day of the Week'},
    text='Total_Incidents'  # Add text labels for the values
)

fig.update_traces(
    textposition='outside'  # Position the text above the bars
)

fig.update_layout(
    xaxis_title='Day of the Week',
    yaxis_title='Number of Mass Shootings',
    yaxis=dict(range=[0, max_value * 1.1]),  # Expand the y-axis range to 110% of max value
    bargap=0.2,
    margin=dict(t=50),  # Adjust the top margin to allow space for titles and labels
)

fig.show()
```

The bar graph shows the distribution of mass shootings by the day of the week, with Sunday clearly standing out as the day with the highest number of incidents (202), followed by Saturday with 154 incidents. In contrast, the weekdays generally see fewer mass shootings, with Thursday having the least (38). This trend suggests that weekends, particularly Sunday, experience a significant spike in mass shooting incidents, potentially indicating behavioral or situational factors that lead to increased violence during this time.

#### School Incidents with Victims: Weekly Trends

The following code calculates the total number of school incidents with victims for each day of the week and creates a bar chart for visualization.

Comparison to Mass Shooting Results:
When comparing the trends between school incidents with victims and mass shootings, notable differences emerge. While Friday sees the most school incidents, Sunday stands out as the day with the highest number of mass shootings. In contrast, Sunday records the fewest school-related incidents, highlighting a stark difference in activity between these datasets.
```python
import pandas as pd

# Summarize mass shootings by state
state_summary = mass_shootings.groupby("State").agg(
    Total_Mass_Shootings=("Incident ID", "count"),
    Total_Killed=("Victims Killed", "sum"),
    Total_Injured=("Victims Injured", "sum")
).reset_index()
```

### Converting State Names to Abbreviations

To use the state-level data in a choropleth heatmap, I will convert the full state names into their standard abbreviations. This ensures compatibility with mapping libraries and makes the data easier to visualize on a map.


```python
# Map of full state names to abbreviations
state_abbreviations = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR", "California": "CA",
    "Colorado": "CO", "Connecticut": "CT", "Delaware": "DE", "Florida": "FL", "Georgia": "GA",
    "Hawaii": "HI", "Idaho": "ID", "Illinois": "IL", "Indiana": "IN", "Iowa": "IA",
    "Kansas": "KS", "Kentucky": "KY", "Louisiana": "LA", "Maine": "ME", "Maryland": "MD",
    "Massachusetts": "MA", "Michigan": "MI", "Minnesota": "MN", "Mississippi": "MS",
    "Missouri": "MO", "Montana": "MT", "Nebraska": "NE", "Nevada": "NV", "New Hampshire": "NH",
    "New Jersey": "NJ", "New Mexico": "NM", "New York": "NY", "North Carolina": "NC",
    "North Dakota": "ND", "Ohio": "OH", "Oklahoma": "OK", "Oregon": "OR", "Pennsylvania": "PA",
    "Rhode Island": "RI", "South Carolina": "SC", "South Dakota": "SD", "Tennessee": "TN",
    "Texas": "TX", "Utah": "UT", "Vermont": "VT", "Virginia": "VA", "Washington": "WA",
    "West Virginia": "WV", "Wisconsin": "WI", "Wyoming": "WY"
}

# Convert full state names to abbreviations
state_summary["State"] = state_summary["State"].map(state_abbreviations)

```

### Heatmap of Mass Shootings by State

Using the summarized data, will run the code below, which I creaste a heatmap to visualize the total number of mass shootings across the United States. States with higher numbers of mass shootings are shown in darker shades, helping to highlight regions with more frequent incidents.


```python
fig = px.choropleth(
    state_summary,
    locations="State",
    locationmode="USA-states",  # Use USA state abbreviations
    color="Total_Mass_Shootings",  # Use Total_Killed or Total_Injured if preferred
    color_continuous_scale="Reds",
    scope="usa",  # Restrict to the USA map
    title="Heatmap of Mass Shootings by State"
)

# Adjust the layout to make the map larger
fig.update_layout(
    title_font_size=24,
    title_x=0.5,  # Center the title
    geo=dict(
        lakecolor="rgb(255, 255, 255)",  # Color lakes white
    ),
    width=1200,  # Adjust width of the figure
    height=800   # Adjust height of the figure
)

fig.show()
```

### Heatmap of Total Victims by State

Running the code below generates a heatmap which focuses on the total number of victims (including fatalities and injuries) from mass shootings in each state. ### Cities/Counties with the Most Victims

To gain more granular insights into the impact of mass shootings, I want to identify which cities/counties have experienced the highest number of fatalities and injuries. This will help reveal specific locations that have been most affected by these incidents, allowing for a deeper understanding of the local impact.

By running the code below, I'll create two bar graphs:

The first will display the cities/counties with the highest fatalities from mass shootings, labeled with the state they are located in.
The second will show the cities/counties with the highest injuries, also labeled with the corresponding state.


```python
import pandas as pd
import plotly.express as px

# Summarize data by City/County and State for total killed and injured
city_summary = mass_shootings.groupby(['City Or County', 'State']).agg(
    Total_Killed=('Victims Killed', 'sum'),
    Total_Injured=('Victims Injured', 'sum')
).reset_index()

# Sort for the top 10 cities/counties with the most killed and most injured
top_killed = city_summary.sort_values(by='Total_Killed', ascending=False).head(10)
top_injured = city_summary.sort_values(by='Total_Injured', ascending=False).head(10)

# Create bar charts for fatalities and injuries
fig_killed = px.bar(
    top_killed,
    x='Total_Killed',
    y='City Or County',
    color='State',
    orientation='h',
    title='Top 10 Cities/Counties with the Most Killed (Mass Shootings)',
    labels={'Total_Killed': 'Total Killed', 'City Or County': 'City/County'},
)

fig_injured = px.bar(
    top_injured,
    x='Total_Injured',
    y='City Or County',
    color='State',
    orientation='h',
    title='Top 10 Cities/Counties with the Most Injured (Mass Shootings)',
    labels={'Total_Injured': 'Total Injured', 'City Or County': 'City/County'},
)

# Customize layout for readability
fig_killed.update_layout(
    xaxis_title='Number of Fatalities',
    yaxis_title='City/County',
    height=600,
    width=800,
)
fig_injured.update_layout(
    xaxis_title='Number of Injuries',
    yaxis_title='City/County',
    height=600,
    width=800,
)

# Display both graphs
fig_killed.show()
fig_injured.show()
```

The bar graphs show that Philadelphia, Pennsylvania, has the highest fatalities (over 25), followed closely by Chicago, Illinois. Chicago, however, leads significantly in injuries, with nearly 150 reported. Cities like Memphis, Tennessee, and Dallas, Texas, appear in the top rankings for both fatalities and injuries, emphasizing their significant share of incidents. Other cities like Baltimore, Maryland, and Washington, D.C., stand out for injuries, while smaller cities such as Lewiston, Maine, and Monterey Park, California, make the top 10 for fatalities, highlighting the varied impact of mass shootings across different city sizes and regions. These findings reflect the geographic diversity of mass shooting impacts in the U.S.
